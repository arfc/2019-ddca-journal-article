```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 10000, 2000)
data1 = np.array([20,9,8,7,7])
data2 = np.array([315791,306520,303438,303438,303438])

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Buffer Size [MW]')
ax1.set_ylabel('No. of Timesteps with Undersupply ', color=color)
ax1.plot(t, data1, color=color,marker='o')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Cumulative Undersupply [GW * mo]', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color,marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('EG01-24: Power Undersupply vs. Power Buffer Size', fontsize=14)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig("../figures/24-sens-buffer", dpi=300, bbox_inches="tight")
plt.show()
```

```python
t = np.arange(0, 10000, 2000)
data1 = np.array([9,6,6,5,5])
data2 = np.array([152517,147466,143166,139083,135083])

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Buffer Size [MW]')
ax1.set_ylabel('No. of Timesteps with Undersupply ', color=color)
ax1.plot(t, data1, color=color,marker='o')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yticks([5,6,7,8,9])

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Cumulative Undersupply [GW * mo]', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color,marker='o')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('EG01-30: Power Undersupply vs. Power Buffer Size', fontsize=14)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig("../figures/30-sens-buffer", dpi=300, bbox_inches="tight")
plt.show()
```
