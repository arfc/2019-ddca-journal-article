```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,201)
D1 = 100-t
D2 = t-99
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.plot(t,D1,label='Preference for LWR (100-t)')
ax.plot(t,D2,label='Preference for SFR (t-99)')
ax.grid(alpha=0.7)
ax.set_xlabel('Timestep [month]', fontsize=18)
ax.set_ylabel('Preference', fontsize=18)
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_ylim(-100,100)
ax.set_xlim(0,200)
ax.axvline(x=100,linestyle=':',color='red',label='Transition Timestep')
ax.legend(loc="upper center", bbox_to_anchor=(0.85, 0.7), fontsize=14)
ax.set_title('Preference for each Reactor Type vs. Time', fontsize=20)
plt.savefig("../figures/prefplot", dpi=300, bbox_inches="tight")
```
