```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0,101)
D1 = 10000*np.ones(101)
D2 = 10000*np.ones(101)
for i in range(40,101):
    D2[i] = 250*i
D3 = 1000*np.sin(np.pi*t/2)+10000
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.plot(t,D2,label='Linearly Increasing')
ax.plot(t,D3,label='Sinusoidal')
ax.plot(t,D1,label='Constant')
ax.grid()
ax.legend(loc="upper center", bbox_to_anchor=(0.12, 1.0), fontsize=14)
ax.set_ylim(0,25000)
ax.set_xlim(0,100)
ax.set_xlabel('Timestep (month)', fontsize=18)
ax.set_ylabel('Power [MW]', fontsize=18)
ax.set_title('Simple Transition Scenarios: Power Demand Curves', fontsize=20)
plt.savefig("../figures/powerplots", dpi=300, bbox_inches="tight")
```
